```python
import pickle
import matplotlib.pyplot as plt
```

```python
times_avg_inception=pickle.load(open('times_avg_inception.pickle','rb'))
times_std_inception=pickle.load(open('times_std_inception.pickle','rb'))
predict_avg_inception=pickle.load(open('predict_avg_inception.pickle','rb'))
predict_std_inception=pickle.load(open('predict_std_inception.pickle','rb'))
times_avg_mobilenet=pickle.load(open('times_avg_mobilenet.pickle','rb'))
times_std_mobilenet=pickle.load(open('times_std_mobilenet.pickle','rb'))
predict_avg_mobilenet=pickle.load(open('predict_avg_mobilenet.pickle','rb'))
predict_std_mobilenet=pickle.load(open('predict_std_mobilenet.pickle','rb'))
times_avg_resnet=pickle.load(open('times_avg_resnet.pickle','rb'))
times_std_resnet=pickle.load(open('times_std_resnet.pickle','rb'))
predict_avg_resnet=pickle.load(open('predict_avg_resnet.pickle','rb'))
predict_std_resnet=pickle.load(open('predict_std_resnet.pickle','rb'))
```

```python
aws_cost_memory=[512,1024,1536,2048,3072]
cost=[0.83,1.67,2.50,3.33,5]
```

#### Warm start experiments

```python
memories=[512,640,1024,1500,1800,2048,2500,3000]
plt.errorbar(memories,times_avg_inception,yerr=times_std_inception,label="Inception Latency")
plt.errorbar(memories,times_avg_resnet,yerr=times_std_resnet,label="RESNET Latency")
plt.errorbar(memories,times_avg_mobilenet,yerr=times_std_mobilenet,label="MobileNet Latency")
plt.plot(aws_cost_memory,cost,label='cost per ms x 10^-8')
plt.title("Time(s) vs Memory(MB)")
plt.legend()
plt.show()
```

```python
plt.errorbar(memories,predict_avg_inception,yerr=predict_std_inception,label="Inception Inference Time")
plt.errorbar(memories,predict_avg_resnet,yerr=predict_std_resnet,label="RESNET Inference time")
plt.errorbar(memories,predict_avg_mobilenet,yerr=predict_std_mobilenet,label="MobileNet Inference time")
plt.plot(aws_cost_memory,cost,label='cost per ms x 10^-8')
plt.title("Time(s) vs Memory(MB)")
plt.legend()
plt.show()
```
